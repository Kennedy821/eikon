```python
# replace this with your credentials
from credentials import my_email, my_password
my_email = my_email
my_password = my_password
```

```python
# note you should upgrade to eikonsai version 0.9.1 to use the jobs module
import eikonsai as eikon
my_api_key =  eikon.utils.get_api_key_from_credentials(email=my_email,
                                         password=my_password)
```

```python
list_of_supported_areas = eikon.utils.list_of_supported_areas()
```

```python
import random
# this is testing the traversal from any area to another
origin_location = random.sample(list_of_supported_areas,1)[0]
destination_location = random.sample([x for x in list_of_supported_areas if x != origin_location],1)[0]
print(f"Origin: {origin_location}")
print(f"Destination: {destination_location}")
```

# Visual area similarity

```python
eikon.similarity.area_similarity(area_1 = origin_location,
                area_2 = destination_location,
                similarity_type = "visual",
                user_api_key = my_api_key)
```

# Descriptive area similarity

```python
eikon.similarity.area_similarity(area_1 = origin_location,
                area_2 = destination_location,
                similarity_type = "descriptive",
                user_api_key = my_api_key)
```

# Combined area similarity

```python
eikon.similarity.area_similarity(area_1 = origin_location,
                area_2 = destination_location,
                similarity_type = "combined",
                user_api_key = my_api_key)
```
