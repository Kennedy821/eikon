# Getting a description of a location

This example will look at Wembley Stadium. We'll collect descriptions of Wembley at the 3 spatial resolutions available:

1) Low
2) Medium
3) High

## Low resolution

```python
location = [51.433727, -0.214443] # wimbledon tennis club
resolution = "low"
eikon.context.get_location_description(lat=location[0],
                                       lon=location[1],
                                       resolution=resolution,
                                    user_api_key=my_api_key)[50:250]
```

# Medium resolution

```python
location = [51.433727, -0.214443] # wimbledon tennis club
resolution = "medium"
eikon.context.get_location_description(lat=location[0],
                                       lon=location[1],
                                       resolution=resolution,
                                    user_api_key=my_api_key)[50:250]
```

```python
location = [51.433727, -0.214443] # wimbledon tennis club
resolution = "high"
eikon.context.get_location_description(lat=location[0],
                                       lon=location[1],
                                       resolution=resolution,
                                    user_api_key=my_api_key)[50:250]
```
