# Getting the similarity of two locations

## Low resolution

```python
location_1 = [51.531143, -0.159893] # regent's park
location_2 = [51.433727, -0.214443] # wimbledon tennis club
resolution = "low"
visual_similarity = eikon.similarity.visual_similarity(location_1_lat_lon_list=location_1,
                             location_2_lat_lon_list=location_2, 
                             resolution=resolution,
                             user_api_key=my_api_key
                            )
print(visual_similarity)
```
